```python
import pandas as pd
```

```python
path1="/content/StatewiseTestingDetails.csv"
path2="/content/covid_19_india.csv"
path3="/content/covid_vaccine_statewise.csv"

stateData=pd.read_csv(path1,parse_dates=['Date']);
stateCovidVaccineData=pd.read_csv(path2,parse_dates=['Date'])
covidData=pd.read_csv(path3,parse_dates=['Updated On'])
```

```python
#Removing Null values rows from dataset
stateData=stateData.dropna(axis=0);
#covidData=covidData.dropna(axis=0);
stateCovidVaccineData=stateCovidVaccineData.dropna(axis=0);
```

```python
stateCovidVaccineData.head()
```

```python
#State wise total covid Cases
stateData.head()
stateData.groupby("State").State.count()
```

```python
#Removed The columns which have all the null values
newCovidData=covidData.drop(columns=['60+ Years(Individuals Vaccinated)','18-44 Years(Individuals Vaccinated)','60+ Years (Doses Administered)','45-60 Years (Doses Administered)','18-44 Years (Doses Administered)','AEFI','Sputnik V (Doses Administered)','45-60 Years(Individuals Vaccinated)','Second Dose Administered','Male (Doses Administered)','Female (Doses Administered)','Transgender (Doses Administered)'])
newCovidData=newCovidData.dropna(axis=0);
newCovidData
#male female transgender vaccinated comparison 
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Covaxin Vs CovidShield Vaccine comparision
cdf=newCovidData.iloc[:,[6,7]].dropna(axis=0)
plt.figure(figsize=(16, 8))
plt.title("Comprisons")
sns.lineplot(data=cdf)
plt.ylabel("Vaccines")
```

##**OBSERVATION**-Covaxin is the majorly used vaccine all over India as compared to covidshiel.:

```python
#Cured Vs death comparition
scvdf=stateCovidVaccineData.iloc[:,[6,7]]
plt.figure(figsize=(16, 8))
plt.title("Comprisons")
sns.lineplot(data=scvdf)
plt.ylabel("people dies/cured")
```

##**OBSERVATION**-Cured vs death comparision cured rate is much greater than death rate in India which specifies India is good as compared to other countries while fighting with COVID-19.

```python
#//Comparison b/w male vs female vaccinated
cdf=newCovidData.iloc[:,[8,9,10]].dropna(axis=0)
plt.figure(figsize=(16, 8))
plt.title("Comparisons")
sns.lineplot(data=cdf)
plt.ylabel("Vaccinated")
```

##**OBSERVATION**-Male are dominant in population so they are in larger numbers of vaccinated as compared to women.

```python
#Total Doses administrated across different states
dwi = covidData.set_index("State")
dwi = dwi.drop("India")
ncd=dwi.reset_index()
ncd=ncd.iloc[:,[0,1,2]];
plt.figure(figsize=(12,9))
sns.barplot(x=ncd['Total Doses Administered'],y=ncd.State)
```

##**OBSERVATION**-From the above visulization Maharashtra holding at first position in Vaccine Doses administered followed by Uttar Pradesh and Gujarat.

```python
#stateWise Total covid cases
nsd=stateData.iloc[:,[0,1,2]];
plt.figure(figsize=(12,12))
plt.title("Total Samples")
sns.barplot(x=nsd['TotalSamples'],y=nsd.State)
```

##**OBSERVATION**-From the above visulization jharkhand is having more number of Positive and Negative cases followed by Maharashtra and Andhra Pradesh.

```python
#statewise comparision b/w positive and negative cases
plt.figure(figsize=(20,15))
plt.title("Negative vs Positive cases on States")
sns.scatterplot(x=stateData.Negative,y=stateData.Positive,hue=stateData.State)
```

##**OBSERVATION**-Above variations shows the variation in negative and positive cases across the different states Mumbai having much more Positive cases rate then negative rate.Some of the states having flat varition showing minor positive rate(less no of covid cases).

```python
#Comparision between different age groups for Doses Administered
cd=covidData.iloc[:,[14,15,16,17,18,16]]
plt.figure(figsize=(18,12))
plt.title("Comprisons")
sns.lineplot(data=cd)
plt.ylabel("Doses Administered")
```

##**OBSERVATION**-From the above graph we can see that the age group between 18-44 years are mostly more affective in Doses administration than others.

```python
#First Dose vs Second Dose
data_with_index = covidData.set_index("State")
data_with_index = data_with_index.drop("India")
cd=data_with_index.reset_index()
cd=cd.iloc[:,[0,5,6]]
plt.figure(figsize=(16,12))
plt.title("First Dose vs Second Dose on States")
sns.scatterplot(x=cd['First Dose Administered'],y=cd['Second Dose Administered'],hue=cd.State)
```

##**OBSERVATION**-We can visualize that some states are having low First Dose administred and Second Dose administred is high.And some states have high first dose administred and second dose low and some of the states having good doses administration of both first and second dose.
